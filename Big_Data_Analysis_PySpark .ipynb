```python
from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .appName("Big Data Analysis Internship Task") \
    .getOrCreate()

print("Spark Session Created Successfully")
```

```python
import pandas as pd

url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/titanic.csv"
pdf = pd.read_csv(url)

pdf.head()
```

```python
data = spark.createDataFrame(pdf)
data.show(5)
```

```python
print("Rows:", data.count())
print("Columns:", len(data.columns))
```

```python
data.printSchema()
```

```python
data.groupBy("survived").count().show()
```

```python
data.selectExpr("avg(age)").show()
```

**Insights from Big Data Analysis using PySpark**

1. PySpark was used to process a large dataset efficiently in a distributed environment.

2. The dataset contained multiple columns with different data types, which were identified using schema analysis.

3. Aggregation operations such as groupBy and average were used to derive meaningful insights.

4. Survival distribution analysis provided an overview of outcome patterns in the dataset.

5. This task demonstrated the scalability of PySpark for handling large datasets.
